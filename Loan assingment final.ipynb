```python
x=df1.iloc[:,:11]
y=df1.iloc[:,-1]
```

```python
x_dummies = pd.get_dummies(x, drop_first=True)
# here i have used get dummies method to covert catagorical data, i am not using one hot encoder as needs time to understand :)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_dummies)
x_scaled = scaler.transform(x_dummies)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2)
```

```python
# i have used descision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)
```

```python
dt.score(x_test,y_test) # my model is over fitting and test also have 100% acquracy which tells some thing is wrong
```

```python
# i am checking whether it same for other classifiers also or not
from sklearn.linear_model import LogisticRegression
lg= LogisticRegression()
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test) # same propblem
```

```python
# with another one
from sklearn.svm import SVC
svc = SVC(kernel = 'linear', random_state = 0)
svc.fit(x_train, y_train)
```

```python
svc.score(x_test,y_test)
```

```python
# by the way i am using ensemble learning although my previous classifiers are overfitting
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=20)
rf.fit(x_train,y_train)
```

```python
from sklearn.ensemble import BaggingClassifier
bc = BaggingClassifier(DecisionTreeClassifier(), max_samples= 0.5, max_features = 1.0, n_estimators = 20)
bc.fit(x_train,y_train)
```

```python
# boosting ensemble 
from sklearn.ensemble import AdaBoostClassifier
adb = AdaBoostClassifier(DecisionTreeClassifier(),n_estimators = 5, learning_rate = 1)
adb.fit(x_train,y_train)
```

```python
# here i am making my own ensemble by using votting
from sklearn.ensemble import VotingClassifier
evc = VotingClassifier( estimators= [('lg',lg),('svc',svc)], voting = 'hard')
evc.fit(x_train,y_train)
evc.score(x_test, y_test)
# i have used two classification method logistic and vector machin to make my ensamble 
```
